### 代码绘制导函数图像

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
```

```python
print(x)
```

```python
print(y)
```

#### 画出函数图像

```python
plt.plot(x,y)
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
```

```python
#一阶导数
y1 = np.cos(x)
plt.plot(x,y1, linestyle='--', color='r')
plt.xlabel("x")
plt.ylabel("y1")
plt.title("y = cos(x)")
```

```python
plt.plot(x,y)
plt.plot(x,y1, linestyle='--', color='r')
plt.xlabel("x")
plt.ylabel("y")
plt.title("y = sin(x)")
```
