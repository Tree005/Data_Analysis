## 1.ndarray的特性
### 多维性

```python
import numpy
```

```python
arr = numpy.array(3)#零维数组
print(f"{arr}的维度为：",numpy.ndim(arr))
```

```python
arr = numpy.array([1,2,3])#1维数组
print(f"{arr}的维度为：",numpy.ndim(arr))
```

```python
arr = numpy.array([[4,5,6],[1,2,3]])#1维数组
print(f"{arr}的维度为：",numpy.ndim(arr))
```

### 同质性

```python
arr = numpy.array([1,"hello"])#不同数据类型会被转换成统一的数据类型
print(f"{arr}的数据类型为：",arr.dtype)
```

### 2.ndarray属性

```python
arr = numpy.array(3)
print("数组形状为：",arr.shape)
print("数组维度为：",arr.ndim)
print("元素的个数为：",arr.size)
print("元素的数据类型：",arr.dtype)
```

```python
arr_1 = numpy.array([1,2.5,3])
print("数组形状为：",arr_1.shape)
print("数组维度为：",arr_1.ndim)
print("元素的个数为：",arr_1.size)
print("元素的数据类型：",arr_1.dtype)
```

```python
arr_2 = numpy.array([[1,2,3],[4,5,6]])
print("数组形状为：",arr_2.shape)
print("数组维度为：",arr_2.ndim)
print("元素的个数为：",arr_2.size)
print("元素的数据类型：",arr_2.dtype)
print("元素的转置：",arr_2.T)
```
