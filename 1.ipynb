## 1.ndarray的特性
### 多维性

```python
import numpy
```

```python
arr = numpy.array(3)#零维数组
print(f"{arr}的维度为：",numpy.ndim(arr))
```

```python
arr = numpy.array([1,2,3])#1维数组
print(f"{arr}的维度为：",numpy.ndim(arr))
```

```python
arr = numpy.array([[4,5,6],[1,2,3]])#2维数组
print(f"{arr}的维度为：",numpy.ndim(arr))
```

### 同质性

```python
arr = numpy.array([1,"hello"])#不同数据类型会被转换成统一的数据类型
print(f"{arr}的数据类型为：",arr.dtype)
```

### 2.ndarray属性

```python
arr = numpy.array(3)
print("数组形状为：",arr.shape)
print("数组维度为：",arr.ndim)
print("元素的个数为：",arr.size)
print("元素的数据类型：",arr.dtype)
```

```python
arr_1 = numpy.array([1,2.5,3])
print("数组形状为：",arr_1.shape)
print("数组维度为：",arr_1.ndim)
print("元素的个数为：",arr_1.size)
print("元素的数据类型：",arr_1.dtype)
```

```python
arr_2 = numpy.array([[1,2,3],[4,5,6]])
print("数组形状为：",arr_2.shape)
print("数组维度为：",arr_2.ndim)
print("元素的个数为：",arr_2.size)
print("元素的数据类型：",arr_2.dtype)
print("元素的转置：",arr_2.T)
```

### 3.ndarray的创建

```python
# 预定义形状

arr_zeros = numpy.zeros((2,3),dtype=int)
print("全零二维三列数组：\n",arr_zeros)
```

```python
arr_ones = numpy.ones((2,3))
print("全一二维数组：\n",arr_ones)
```

```python
arr_empty = numpy.empty((3,3)) 
print("空二维数组：\n",arr_empty)
```

```python
arr_pull = numpy.full((2,3),7)
print("全7二维数组：\n",arr_pull)
```

```python
# 根据已有数组形状创建新数组
arr_empty_like = numpy.empty_like(arr_pull)
print("根据arr_pull形状创建的空数组：\n",arr_empty_like)
print("--- ----------------------------------------------")
arr_ones_like = numpy.ones_like(arr_pull)
print("根据arr_pull形状创建的全一数组：\n",arr_ones_like)
print("--- ----------------------------------------------")
arr_zeros_like = numpy.zeros_like(arr_pull)
print("根据arr_pull形状创建的全零数组：\n",arr_zeros_like)
print("--- ----------------------------------------------")
arr_full_like = numpy.full_like(arr_pull,9)
print("根据arr_pull形状创建的全9数组：\n",arr_full_like)
```

```python
#
arr_a = numpy.arange(1,10,2)
print("arange创建的数组：\n",arr_a)
```

```python
#等间隔数列
arr_b = numpy.linspace(0,100,5)
print("linspace创建的数组：\n",arr_b)
print("--- ----------------------------------------------")
arr_b_arange = numpy.arange(0,101,25)
print("arange创建的等间隔数组：\n",arr_b_arange)
```

```python
#对数等差数列
arr_c = numpy.logspace(0,3,2,base = 2)
print("logspace创建的数组：\n",arr_c)
```

```python
#特殊矩阵
#单位矩阵:对角线为1，其余元素为0的方阵
arr_eye = numpy.eye(3,4,dtype=int)
print("单位矩阵：\n",arr_eye)
```

```python
# 对角矩阵：主对角线上元素非零，其余元素为零的方阵
arr_diag = numpy.diag([1,2,3,4])
print("对角矩阵：\n",arr_diag)
```

```python
# 随机数矩阵
# 生成0-1之间均匀分布的随机数矩阵
arr_rand = numpy.random.rand(2,3)#二行三列
print("随机数矩阵：\n",arr_rand)
```

```python
# 生成指定区间的随机浮点数
arr_uniform = numpy.random.uniform(1,10,(2,3))#1-10 二行三列
print("均匀分布随机数矩阵：\n",arr_uniform)
```

```python
# 生成指定区间的随机整数
arr_randint = numpy.random.randint(1,10,(2,3))#1-10 二行三列
print("随机整数矩阵：\n",arr_randint)
```

```python
# 生成随机数（正态分布）-3 -  3
arr_randn = numpy.random.randn(2,3) #二行三列
print("正态分布随机数矩阵：\n",arr_randn)
```

```python
# 生成随机种子
numpy.random.seed(20)
arr_randint = numpy.random.randint(1,10,(2,3))#1-10 二行三列
print("设置随机种子后的随机整数矩阵：\n",arr_randint)
```

### 5.索引与切片

```python
arr = numpy.random.randint(1,100,30)
print("随机生成的整数数组为：\n",arr)
```

```python
# 对一维数组进行切片 支持逻辑运算符
print(arr[:])#获取全部元素
print("-- ----------------------------------------------")
print(arr[slice(2,20,1)])
print("--- ----------------------------------------------")
print(arr[0:5])#从索引0到5(不包含5)步长默认为1
print("--- ----------------------------------------------")
print(arr[0:10:2])#从索引0到10(不包含10)步长为2
print("--- ----------------------------------------------")
print(arr[arr>10])#大于10的元素
```

```python
arr = numpy.random.randint(1,100,(3,4))
print("随机生成的二维整数数组为：\n",arr)
```

```python
print(arr[1,2])#获取第2行第3列元素
print("--- ----------------------------------------------")
print(arr[:,:])#获取所有行所有列元素
print("--- ----------------------------------------------")
print(arr[0:3,2:4])#获取第1到3行，第3到4列元素
print("--- ----------------------------------------------")
print(arr[2][arr[2] > 50])#获取第3行大于50的元素
print("--- ----------------------------------------------")
```
